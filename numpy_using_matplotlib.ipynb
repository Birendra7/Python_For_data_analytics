ploting trigonometric curve

```python
x_sin = np.arange(0, 2*np.pi,0.1)
y_sin = np.sin(x_sin)

print(y_sin)
```

```python
plt.figure(figsize=(6,6))
plt.plot(x_sin, y_sin, 'y-')
plt.title('Sin Curve')
plt.show
```

```python
x_cos = np.arange(0, 2*np.pi, 0.1)
y_cos = np.cos(x_cos)

plt.figure(figsize=(6,6))
plt.plot(x_cos, y_cos, 'y-')
plt.title('Cos Curve')
plt.show()
```

```python
x_tan = np.arange(0, 2*np.pi, 0.1)
y_tan = np.tan(x_tan)

plt.figure(figsize=(6,6))
plt.title('tan curve')
plt.plot(x_tan, y_tan, 'r-')
plt.show()
```

```python
x_cot = np.arange(0, 2*np.pi, 0.1)
y_cot = 1/np.tan(x_cot)

plt.figure(figsize=(6,6))
plt.plot(x_cot, y_cot, 'h-')
plt.title('cot curve')
plt.show()
```

subplot

```python
plt.figure(figsize=(6,6))
plt.title('subplot')

plt.subplot(2,2,1)
plt.plot(x_sin, y_sin, 'r-')
plt.title('sin curve')

plt.subplot(2,2,2)
plt.plot(x_cos, y_cos, 'b-')
plt.title('cos curve')

plt.subplot(2,2,3)
plt.plot(x_tan, y_tan, 'y-')
plt.title('tan curve')

plt.subplot(2,2,4)
plt.plot(x_cot, y_cot, 'w-')
plt.title('cot curve')
```
